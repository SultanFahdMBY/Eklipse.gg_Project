```python
import pandas as pd

# Load the datasets
gamesession_df = pd.read_csv("/mnt/data/gamesession.csv")
clips_df = pd.read_csv("/mnt/data/clips.csv")

# Filter only gamesessions that have been submitted
submitted_sessions = gamesession_df[gamesession_df['submited_date'].notnull()]

# Count total submitted sessions
total_submitted = submitted_sessions['id'].nunique()

# Join submitted sessions with clips on gamesession ID
merged = pd.merge(clips_df, submitted_sessions, left_on='gamesession_id', right_on='id', how='inner')

# Count how many unique gamesession IDs from submitted ones have at least one clip
converted_sessions = merged['gamesession_id'].nunique()

# Calculate conversion rate
conversion_rate = converted_sessions / total_submitted if total_submitted else 0

# Output summary
{
    "Total Submitted Gamesessions": total_submitted,
    "Gamesessions with Clips": converted_sessions,
    "Clip Conversion Rate (%)": round(conversion_rate * 100, 2)
}

```

```python
# --- 1. PREPARATION ---

# Convert datetime columns
gamesession_df['submited_date'] = pd.to_datetime(gamesession_df['submited_date'], errors='coerce')
clips_df['created_at'] = pd.to_datetime(clips_df['created_at'], errors='coerce')

# Convert duration from microseconds to seconds
gamesession_df['duration_sec'] = gamesession_df['duration'] / 1_000_000

# --- 2. RFM METRICS ---

# Recency: Last activity (submit or clip created)
last_stream = gamesession_df.groupby('user_id')['submited_date'].max()
last_clip = clips_df.groupby('user_id')['created_at'].max()
last_activity = pd.concat([last_stream, last_clip], axis=1).max(axis=1)

# Frequency: Total number of activities (stream submissions + clip creations)
stream_count = gamesession_df.groupby('user_id')['id'].count()
clip_count = clips_df.groupby('user_id')['id'].count()
total_activity = stream_count.add(clip_count, fill_value=0)

# Monetary: Total stream duration (as proxy of content produced)
total_duration = gamesession_df.groupby('user_id')['duration_sec'].sum()

# --- 3. COMBINE RFM TABLE ---

rfm_df = pd.DataFrame({
    'last_activity': last_activity,
    'frequency': total_activity,
    'monetary_duration': total_duration
})

# Recency in days (based on latest activity date)
reference_date = rfm_df['last_activity'].max()
rfm_df['recency_days'] = (reference_date - rfm_df['last_activity']).dt.days
rfm_df.drop(columns='last_activity', inplace=True)

# --- 4. FILTER PREMIUM USERS ONLY ---

# Ambil hanya user_id yang ada di premium
premium_users = premium_df['user_id'].unique()
rfm_premium = rfm_df[rfm_df.index.isin(premium_users)]

# Output top 10
rfm_premium
```

```python
# Convert qcut ke string agar bisa fillna dengan '0'
rfm_df['R_score'] = pd.qcut(rfm_df['recency_days'], 5, labels=[5, 4, 3, 2, 1]).astype(str)
rfm_df['F_score'] = pd.qcut(rfm_df['frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5]).astype(str)
rfm_df['M_score'] = pd.qcut(rfm_df['monetary_duration'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5]).astype(str)

# Isi NaN dengan '0' string
rfm_df[['R_score', 'F_score', 'M_score']] = rfm_df[['R_score', 'F_score', 'M_score']].fillna('0')

# Buat kolom gabungan
rfm_df['RFM_score'] = (
    rfm_df['R_score'] +
    rfm_df['F_score'] +
    rfm_df['M_score']
)
```

```python
rfm_counts = rfm_df['RFM_score'].value_counts().reset_index()
rfm_counts.columns = ['RFM_score', 'User Count']
sns.barplot(data=rfm_counts.head(10), x='RFM_score', y='User Count')
plt.title("Top 10 RFM Scores by User Count")
plt.show()
```

```python
# Pastikan semua skor string 'nan' diganti dengan angka, misalnya 0
rfm_df['R_score'] = rfm_df['R_score'].replace('nan', 0).astype(int)
rfm_df['F_score'] = rfm_df['F_score'].replace('nan', 0).astype(int)
rfm_df['M_score'] = rfm_df['M_score'].replace('nan', 0).astype(int)
```

```python
def segment_user(row):
    r = row['R_score']
    f = row['F_score']
    m = row['M_score']

    if r >= 4 and f >= 4 and m >= 4:
        return "Champion"
    elif r >= 4 and f >= 3:
        return "Loyal Customer"
    elif r >= 3 and f >= 3 and m <= 2:
        return "Big Spender, Needs Attention"
    elif r <= 2 and f >= 4:
        return "At Risk"
    elif r >= 3 and f <= 2:
        return "Recent Customer"
    elif r <= 2 and f <= 2:
        return "Lost"
    elif r >= 2 and f >= 3 and m <= 3:
        return "Potential Loyalist"
    elif r <= 3 and f == 5:
        return "Frequent but Aging"
    else:
        return "Others"
```

```python
rfm_df['segment'] = rfm_df.apply(segment_user, axis=1)
```
