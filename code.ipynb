```python
import os
import pandas as pd
from google import genai
from tqdm import tqdm
import time

# Set up the Gemini client
client = genai.Client()  # Assumes API key already set in environment variable

# Load CSV
df = pd.read_csv("Game Thumbnail.csv")

# Prepare new columns
genres = []
descriptions = []
modes = []

# Define helper function
def get_response(prompt, retries=3, delay=1.5):
    for attempt in range(retries):
        try:
            response = model.generate_content(prompt)
            return response.text.strip()
        except Exception as e:
            if "RESOURCE_EXHAUSTED" in str(e):
                print("⚠️ Rate limit hit. Waiting...")
                time.sleep(5)
            else:
                return f"ERROR: {e}"
    return "ERROR: Retry failed"

# Loop through rows
for title in tqdm(df["game_title"]):
    # Genre
    genre_prompt = f"What is the genre of the video game '{title}'? Respond with only one word."
    genre = get_response(genre_prompt).split()[0]  # Ambil 1 kata pertama

    # Short Description
    desc_prompt = f"Write a short, concise description (under 30 words) for a video game titled '{title}'."
    desc = get_response(desc_prompt)

    # Player Mode
    mode_prompt = (
        f"Is the video game '{title}' primarily 'Singleplayer', 'Multiplayer', or 'Both'? "
        f"Respond with one of these exact words."
    )
    mode = get_response(mode_prompt).split()[0].capitalize()

    genres.append(genre)
    descriptions.append(desc)
    modes.append(mode)

# Tambahkan kolom ke DataFrame
df["genre"] = genres
df["short_description"] = descriptions
df["player_mode"] = modes

# Simpan ke file baru
df.to_csv("Game_Thumbnail_Enhanced.csv", index=False)
print("✅ Saved as Game_Thumbnail_Enhanced.csv")
```
